## tic-tac-toe game

##### Number the houses so that each user can fill it by choosing its number.
##### If the number is out of range, the player's turn will burn.

```python
print("""| 1 | 2 | 3 |
-------------
| 4 | 5 | 6 |
-------------
| 7 | 8 | 9 |\n""")

print("""How to play: Enter your desired house number so that it can be filled for you if it is empty.
If you enter the wrong number, you will lose your turn.""") #Show how to play

print("==========================================\n")
turn = True #For unlimited loops
home_list = [0,' ',' ',' ',' ',' ',' ',' ',' ',' ']
users_lst = ['a','b']

while turn:
    for user in users_lst:
        
        print("===============")
        print(f"| user {user} turn |") # Display "Which player's turn is it?"
        print("===============")
        
        unit = int(input("Enter the unit number you want: "))

        if home_list[unit] == ' ':
            
            if user == 'a':
                home_list[unit] = 'a'

            else:
                home_list[unit] = 'b'

            print(f"""| {home_list[1]} | {home_list[2]} | {home_list[3]} |
-------------
| {home_list[4]} | {home_list[5]} | {home_list[6]} |
-------------
| {home_list[7]} | {home_list[8]} | {home_list[9]} |\n""") #Show the updated table

        else:
            break

        if (home_list[1] == home_list[4] == home_list[7] == 'a' or
        home_list[2] == home_list[5] == home_list[8] == 'a' or
        home_list[3] == home_list[6] == home_list[9] == 'a' or
        home_list[1] == home_list[2] == home_list[3] == 'a' or
        home_list[4] == home_list[5] == home_list[6] == 'a' or
        home_list[7] == home_list[8] == home_list[9] == 'a' or
        home_list[1] == home_list[5] == home_list[9] == 'a' or
        home_list[3] == home_list[5] == home_list[7] == 'a') : # user A win

            print("=================")
            print("|               |")
            print("|  User A Win!  |")
            print("|               |")
            print("=================")
            turn = False
            break

        elif (home_list[1] == home_list[4] == home_list[7] == 'b' or
        home_list[2] == home_list[5] == home_list[8] == 'b' or
        home_list[3] == home_list[6] == home_list[9] == 'b' or
        home_list[1] == home_list[2] == home_list[3] == 'b' or
        home_list[4] == home_list[5] == home_list[6] == 'b' or
        home_list[7] == home_list[8] == home_list[9] == 'b' or
        home_list[1] == home_list[5] == home_list[9] == 'b' or
        home_list[3] == home_list[5] == home_list[7] == 'b') : # user B win

            print("=================")
            print("|               |")
            print("|  User B Win!  |")
            print("|               |")
            print("=================")
            turn = False
            break

        elif (home_list[1] != ' ' and home_list[2] != ' ' and home_list[3] != ' ' and
        home_list[4] != ' ' and home_list[5] != ' ' and home_list[6] != ' ' and
        home_list[7] != ' ' and home_list[8] != ' ' and home_list[9] != ' ') : # Draw
        
            print("=================")
            print("|               |")
            print("|  Draw match!  |")
            print("|               |")
            print("=================")
            turn = False
            break

```
