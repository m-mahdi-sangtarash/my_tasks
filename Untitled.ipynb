## Rock, Paper, Scissor Game

##### Write the code of the two-player game (real person with computer) using dictionary.

1.Rock should have value 1, paper value 2 and scissors value 3.

2.The game will be played in 10 rounds and the result will be announced at the end.

```python
import random as rnd

client_module = {"rock":3,"paper":2,"scissor":1}
client_score=0
system_score=0

for i in range (10):
    
    c_module_give = input( "please enter a module from( rock, paper, scissor ):" )
    client_module_val = client_module[c_module_give]
    system_module = rnd.randint(1,3)

#sistem win
    if ( client_module_val == 2 and system_module == 1 ) or ( client_module_val == 1 and system_module == 3 ) or ( client_module_val == 3 and system_module == 2 ):
        
        system_score += 1
        print('=============================================')
        print('| system win this round!                    |')
        print('|                                           |')
        print(f'|result: client score: {client_score} -- system score: {system_score} |')
        print('=============================================')
#client win
    elif (client_module_val == 1 and system_module == 2) or (client_module_val == 3 and system_module == 1) or (client_module_val == 2 and system_module ==3 ):
        
        client_score += 1
        print('=============================================')
        print('| you win this round!                       |')
        print('|                                           |')
        print(f'|result: client score: {client_score} -- system score: {system_score} |')
        print('=============================================')
#equel
    else :
        print('=============================================')
        print('| equel!                                    |')
        print('|                                           |')
        print(f'|result: client score: {client_score} -- system score: {system_score} |')
        print('=============================================')

if client_score > system_score:
    
    print('final result : you win')

elif system_score > client_score:
    
    print('final result : system win')
else:
    
    print('final result : draw Match')
```
