## Goal or nonsense game

##### In this game, the system puts the stone in one of its hands without telling the user. The user must guess in which hand the system has hidden the stone. If he guesses the right hand, he should enter the value of r, and if he guesses the left hand, he should enter the value of l.

1.If the user guesses correctly, one point is added to the user's points, and if he guesses wrong, one point is added to the system's points.

2.Continue the game for 10 rounds and display the result.

```python
import random as rnd

# For ease of work, we convert the random number that the system has chosen for itself into letters
system_module_item = {1: "r", 2: "l"} 

sys_score = 0
client_score = 0

for turn in range (10):
    
    # The system generates a random number and places the stone in one of its hands.
    
    sys_module = rnd.randint(1,2)
    sys_module_compile = system_module_item[sys_module]
    
    # We ask the user to guess which hand the stone is in the system
    
    client_guess = str.lower(input("Which one of my hands do you think the stone is in? (Left: l , Right: r): "))
    
    if client_guess == sys_module_compile:
        client_score += 1
        print("==========================================")
        print("|                                        |")
        print("| You won this round! Keep it up hero!   |")
        print("|                                        |")
        print(f"|   your score | {client_score} -- {sys_score} | system score   |" )
        print("|                                        |")
        print("==========================================")
    
    else:
        sys_score += 1
        print("=============================================")
        print("|                                           |")
        print("| This system won this round. Try harder!   |")
        print("|                                           |")
        print(f"|   your score | {client_score} -- {sys_score} | system score      |" )
        print("|                                           |")
        print("=============================================")
        
        
if client_score > sys_score:
    
    print("=============================================")
    print("|               final result                |")
    print("|             You won the game!             |")
    print("|                                           |")
    print(f"|   your score | {client_score} -- {sys_score} | system score   |" )
    print("|                                           |")
    print("=============================================")

elif client_score < sys_score:
    
    print("=============================================")
    print("|               final result                |")
    print("|          The system won the game!         |")
    print("|                                           |")
    print(f"|   your score | {client_score} -- {sys_score} | system score   |" )
    print("|                                           |")
    print("=============================================")
    
else:
    
    print("=============================================")
    print("|                final result               |")
    print("|            The game equalised!            |")
    print("|                                           |")
    print(f"|   your score | {client_score} -- {sys_score} | system score   |" )
    print("|                                           |")
    print("=============================================")
```
