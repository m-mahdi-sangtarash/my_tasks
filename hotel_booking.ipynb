### Write a code that can reserve the rooms of a hotel (which has a total of 27 rooms) during a month. The hotel rooms are as follows:

        | single | Double | Three |
class A |    3   |    4   |   2   |
class B |    3   |    4   |   2   |
class C |    3   |    4   |   2   |

* You should ask the user what kind of room (class and number of beds) he needs and for how many days, and if it is empty, get the user's information and reserve the room.

```python
a = ["", [[], [], []], [[], [], [], []], [[], []]]
b = ["", [[], [], []], [[], [], [], []], [[], []]]
c = ["", [[], [], []], [[], [], [], []], [[], []]]

while True: # For the unlimited number of this question
    choose_class = (str(input("please select your room class (A, B, C): "))).upper() # Select the room class desired by the user

    if choose_class not in ["A", "B", "C"]: # Wrong input warning to the user
        print("\n----Your chosen class does not exist!---\n") 
        continue
        
    else: # Display the user's choice and continue the route
        print("\n=================================")
        print(f"|  your room class   :       {choose_class}  |")
        print("=================================")
        
        # This section is written in this way so that if we add a new room in the future, the code will not be disturbed.
        
        if choose_class == "A": 
            bed_range = (len(a)) - 1
            
        elif choose_class == "B":
            bed_range = (len(b)) - 1
            
        elif choose_class == "C":
            bed_range = (len(c)) - 1

        while True: # For the unlimited number of this question
            choose_number_bed = int(input(f"\nPlease select the number of beds you want (1, ...,{bed_range}): "))

            if choose_number_bed not in range(1,bed_range + 1):
                print("\n---The number of beds you selected does not exist!---\n")
                continue
            
            else:
                # The user chooses how many beds his room has
                if choose_class == "A":
                    room_range = (len(a[choose_number_bed]))
                    
                elif choose_class == "B":
                    room_range = (len(b[choose_number_bed]))
                    
                elif choose_class == "C":
                    room_range = (len(a[choose_number_bed]))
                    
                print("\n=================================")
                print(f"|  your room class   :       {choose_class}  |")
                print(f"|  bed nubmer        :       {choose_number_bed}  |")
                print("=================================")
                
                while True: # For the unlimited number of this question
                    
                    choose_room = int(input(f"\nplease choose your room (1, ..., {room_range}): "))
                    # Select the room number desired by the user
                
                    if choose_room not in range (1, room_range + 1):
                        print("\n---Your selected room number does not exist!---\n")
                        continue
                    
                    else:
                        # Display the list of reserved dates of the room selected by the user to avoid time interference.
                        if choose_class == "A":
                            print("\nreservation list:\n")
                            print(a[choose_number_bed][choose_room - 1])
                            reserve_list = a[choose_number_bed][choose_room - 1]
                            
                        elif choose_class == "B":
                            print("\nreservation list:\n")
                            print(b[choose_number_bed][choose_room - 1])
                            reserve_list = b[choose_number_bed][choose_room - 1]
                            
                        elif choose_class == "C":
                            print("\nreservation list:\n")
                            print(c[choose_number_bed][choose_room - 1])
                            reserve_list = c[choose_number_bed][choose_room - 1]
                            
                        print("\n=================================")
                        print(f"|  your room class   :       {choose_class}  |")
                        print(f"|  bed nubmer        :       {choose_number_bed}  |")
                        print(f"|  room number       :       {choose_room}  |")
                        print("=================================")
                            
                        choose_day = int(input("\nEnter the desired day: "))
                        
                        if choose_day not in reserve_list:
                            reservation_period = int(input("\nPlease specify the duration of your stay: "))
                            
                            for time in range (1, reservation_period + 1):
                                # We check if the days selected by the user for stay conflict with the reservation list?
                                day_res = choose_day + time
                                
                                
                                if day_res not in reserve_list and day_res <= 30:
                                    continue
                                    
                                elif day_res > 30:
                                    print("\n---Your time is out of time!---\n")
                                    break
                                    
                                else:
                                    print("\n---The duration of your time with the reservation list interferes with time!---\n")
                                    break
                            
                            for time in range (0, reservation_period):
                                day_res = choose_day + time
                                if choose_class == "A":
                                    a[choose_number_bed][choose_room - 1].append(day_res)
                                elif choose_class == "B":
                                    b[choose_number_bed][choose_room - 1].append(day_res)
                                elif choose_class == "C":
                                    c[choose_number_bed][choose_room - 1].append(day_res)
                            
                            # Show reservation request record
                            print("===================")
                            print("|  days reserve   |")
                            print("===================\n")
                            
                            # Show reservation information
                            print("\n===============================")
                            print(f"|  your room class   :     {choose_class}  |")
                            print(f"|  bed nubmer        :     {choose_number_bed}  |")
                            print(f"|  room number       :     {choose_room}  |")
                            print(f"|  date              :     {choose_day}  |")
                            print(f"|  reservation period:     {reservation_period}  |")
                            print("===============================")
                                    
                    break
                    
            break    # Go first to register a new application
                        
```
