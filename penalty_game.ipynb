## Penalty game

##### First, you take a penalty and the system must guess where your kick is, then it is the system's turn to kick and you guess where the kick is.

1.The parts you can hit are marked with numbers. Enter your desired section number and try to score!
2.When taking a penalty, the numbering system of the sections is in the same order. Enter the number of the section where you think the system will hit and try to block the penalty!
3.If you enter a number outside the specified range, you will lose 100% of the penalty!
4.If no one wins in the first 5 shots, you continue one round until someone wins!

```python
import random as rnd

print("choose a number\n")
print("=============================")
print("|| [1]       [2]       [3] ||")
print("||                         ||")
print("|| [4]       [5]       [6] ||")
print("||                         ||")
print("|| [7]       [8]       [9] ||")

turn = 5
kicker_lst = ["user", "system"]
user_score = 0
system_score = 0
user_status = ""
system_status = ""

while turn:

    for kicker in kicker_lst:
        if kicker == "user":
            user_kick = int(input("\nyour choose zone: "))
            system_gk = rnd.randint(1,9)

            if user_kick == system_gk:
                user_status += "X-"
                print("\n=============================")
                print("||                         ||")
                print("||                         ||")
                print("||      lose penalty!      ||")
                print("||                         ||")
                print("||                         ||\n")
                print(f"your score| {user_score} -- {system_score} |system_Score")
                print("=================================")
                print("you   : ", user_status)
                print("system: ", system_status)

            elif user_kick != system_gk and user_kick >= 1:
                user_score += 1
                user_status += "O-"
                print("\n=============================")
                print("||                         ||")
                print("||                         ||")
                print("||      Goalllll!!!!       ||")
                print("||                         ||")
                print("||                         ||\n")
                print(f"| System jump : {system_gk} |\n")
                print(f"your score| {user_score} -- {system_score} |system_Score")
                print("=================================")
                print("you   : ", user_status)
                print("system: ", system_status)
                
            elif user_kick != system_gk and user_kick <= 0:
                user_status += "X-"
                print("\n=============================")
                print("||                         ||")
                print("||           X X           ||")
                print("||            O            ||")
                print("||                         ||")
                print("||      out of target      ||\n")
                print(f"your score| {user_score} -- {system_score} |system_Score")
                print("=================================")
                print("you   : ", user_status)
                print("system: ", system_status)

        else :
            user_gk = int(input("\nyour choose zone: "))
            system_kick = rnd.randint(1,9)

            if user_gk == system_kick:
                system_status += "X-"
                print("\n=============================")
                print("||                         ||")
                print("||                         ||")
                print("||  You saved the penalty! ||")
                print("||                         ||")
                print("||                         ||\n")
                print(f"your score| {user_score} -- {system_score} |system_Score")
                print("=================================")
                print("you   : ", user_status)
                print("system: ", system_status)

            else:
                system_score += 1
                system_status += "O-"
                print("\n=============================")
                print("||                         ||")
                print("||                         ||")
                print("||      oh!!Goalll!!!      ||")
                print("||                         ||")
                print("||                         ||\n")
                print(f"| System kick area : {system_kick} |\n")
                print(f"your score| {user_score} -- {system_score} |system_Score")
                print("=================================")
                print("you   : ", user_status)
                print("system: ", system_status)

            turn -= 1

            if user_score == system_score and turn == 0:
                turn += 1

            elif user_score > system_score and turn == 0:
                print("===================")
                print("|                 |")
                print("|     you win!    |")
                print("|                 |")
                print("===================")

            elif user_score < system_score and turn == 0:
                print("===================")
                print("|                 |")
                print("|   system win!   |")
                print("|                 |")
                print("===================")

```
